```python
x = arange(10,40)
y = [3.239, 3.118, 3.004, 2.897, 2.795, 2.700, 2.610, 2.526, 2.446,2.371, 2.300, 2.233, 2.169, 2.110, 2.053, 2.000, 1.950, 1.902,1.857, 1.815, 1.774, 1.736, 1.700, 1.666, 1.634, 1.603, 1.574,1.547, 1.521, 1.496]
Y = arange(1.5,3.2,.001)
```

```python
from scipy.interpolate import interp1d
def Temp(Resistance):
    temperature = interp1d(y[::-1],x[::-1])(Resistance)
    return temperature
```

```python
print Temp(1.574)
print Temp(3.004)
print Temp(2.000)
print Temp(1.75)
```

```python
plot(Y, Temp(Y), linewidth=1.5, color='blue')
ylabel('Temperature (Deg Celcius)')
xlabel('Resistance(MOhms)')
#d = 1.75
plot(1.75, Temp(1.75), 'r.', markersize=10)
annotate('$1.75 M\Omega$, $30.6C^o$',xy=(1.75,30.6315789474),xytext=(2,30.6))
show()
```

Question 3

```python
def finiteDifference(x,y):
    dydx = zeros(y.shape,float)
    dydx[:-1] = diff(y)/diff(x)
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

```python
x = arange(0,10.1, .1) 
y = x**5 
y_prime = 5*x**4 
plot(x,finiteDifference(x,y)) 
plot(x,y_prime) 
show() 
```

The spacing result is very accurate, pretty much identical.

```python
x = arange(0,10.001, .001) 
y = x**5 
y_prime = 5*x**4 
plot(x,finiteDifference(x,y)) 
plot(x,y_prime) 
show() 
```
