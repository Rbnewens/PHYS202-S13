Electric Dipole Potential

\begin{equation}
V(x,y) = \frac{kq}{\sqrt{x^2 + (y-d)^2}} - \frac{kq}{\sqrt{x^2 + (y+d)^2}}
\end{equation}

```python
#make the x-y grid to cover the plane -2<x<2 and -2<y<2
dx = 0.06
dy = 0.06
x,y = meshgrid(arange(-2.,2.,dx),arange(-2.,2.,dy))

x.ndim
y.shape
```

```python
x.ndim
```

```python
k = 1; q = 1; d = 1
V = (k*q/(x**2+(y-d)**2)**(1/2.)) - (k*q/(x**2+(y+d)**2)**(1/2.))


```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x, y, V, rstride=4, cstride=1, linewidth=0,cmap=cm.Greys)
                    
cb = fig.colorbar(p, shrink=0.5)
cset = ax.contour(x, y, V, zdir='z', offset=-40., cmap=cm.coolwarm)
cset = ax.contour(x, y, V, zdir='x', offset=-2., cmap=cm.coolwarm)
#cset = ax.contour(x, y, V, zdir='y', offset=2., cmap=cm.coolwarm)
```

```python
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -6, -8])
contour(x,y,V,levels,cmap=cm.gist_earth)
show()
```

Ex1. Make a 3-D plot of the function
\begin{equation} G(x,y) = \exp(-x^2-y^2) \end{equation} 
over the range -6 < (x,y) < 6 using 51 bins in x and 51 bins in y

```python
dx = 0.05
dy = 0.05
x,y = meshgrid(arange(-6.,6.,dx),arange(-6.,6.,dy))
G = exp(-x**2 - y**2)
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x, y, G, rstride=4, cstride=1, linewidth=0,cmap=cm.Greys)
                    
cb = fig.colorbar(p, shrink=0.5)
cset = ax.contour(x, y, G, zdir='z', offset=-40., cmap=cm.coolwarm)
cset = ax.contour(x, y, G, zdir='x', offset=-2., cmap=cm.coolwarm)
```
