```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import dates
import datetime
data = loadtxt("/nfshome/rbnewens/sunspots.txt")
x = data[:,0]
y = data[:,1]
```

```python
plot(x, y, marker='o', color='red')
xlabel('Months since 1794')
ylabel('Sunspots')
title('Sunspots 1794-2013')
legend('sunspots', loc='lower right')
grid(True)
```

```python
def movingaverage(interval, window_size):
    window= numpy.ones(int(window_size))/float(window_size)
    return numpy.convolve(interval, window, 'same')

x = data[:,0]
y = data[:,1]


plot(x,y, "k.", marker='o', color='red')
y_av = movingaverage(y, 10)
plot(x, y_av,"r", color='black', linewidth=2.0)
xlim(0,1000)
xlabel("Months since Jan 1749.")
ylabel("Sun spots")
title('Sunspots vs Avg Sunspots')
grid(True)
legend(['Sunspots', 'Avg Sunspots'], loc='upper right')
```
