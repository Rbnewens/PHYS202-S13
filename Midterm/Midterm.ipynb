```python
x = np.arange(-pi, pi+0.1, 0.1)
y = (cos(2*x)*cos(x/2))**2
z = (cos(2*x)*cos(x/2))**2
```

```python
plot(x,y, color = 'red', lw=1)
ax = subplot(1,1,1)
plot(x,z, color = 'blue', linestyle='--', lw=1.5)
xlabel(r'$\theta$ (rad)')
ylabel('Intensity')
x_pi = x/np.pi
unit = 0.25
xtick = np.arange(-1, 1+unit, unit)
x_label = [r"$-\pi$", r"$-\frac{\pi}{2}$", r"$-\frac{\pi}{4}$", r"$0$", r"$\frac{\pi}{4}$", r"$\frac{\pi}{2}$", r"$\pi$"]
ax.set_xticks(xtick*np.pi)
ax.set_xticklabels(x_label)
```
