```python
#Weibull Distribution
from numpy import random
f = lambda x: a*(b**(-a))*(x**(a-1))*exp(-(x/b)**a)
   
#analytically    
#a)
CDF = lambda x: 1-exp(-(x/b)**(a))
#b)
Finv = lambda z: b*(-log(1-z))**(1/a)
```

```python
N = 10000
x = arange(0,20,0.001)
z = numpy.random.random_sample(N)
a = 1.5
b = 6
x_samples = Finv(z)

hist(x_samples,50,normed=True,label='sample distribution')
plot(x,f(x),label='True distribution',lw=3,color='r')
plot(x,CDF(x),label='CDF',lw=3,color='k')
xlabel('x')
ylabel('y')

xlim(0,20)
ylim(0,1)

legend()
show()
```
