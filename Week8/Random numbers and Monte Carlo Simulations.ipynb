Example 1: Analytic Inverse CDF Transform

```python
x = arange(0.,pi+0.001,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.
    
plot(x,f(x),label='True distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
```

F-1(z) = cos-1(cos(0)-2z)

```python
#number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)
    
#sampled x values
xsamples = Finv(z);

#how do they complre to the original?
hist(xsamples,50,normed=True,label="Sampled distribution")
plot(x,f(x),label='True distribution',lw=3,color='red')

xlim(0,pi)
ylim(0,0.7)

legend(loc=0)
show()
```

Exercise 1. Exponential function -analytic

```python
x = arange(-10.,10.+0.001,0.01)
f = lambda x: exp(-x)
F = lambda x: -exp(-x)+1
    
plot(x,f(x),label='True distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=3)

xlim(-5.,5.)
ylim(-100.,100.)

legend(loc=0)
show()
```

```python
def distribute1D(x,prob,N): 
    """takes any numerical distribution prob, on the interval defined by the array x, and returns an array of N sampled values that are statisically the same as the input data."""
    Fx = cumsum(prob)/sum(prob) 
    z = numpy.random.random_sample(N) 
    xsamples = x[searchsorted(Fx,z)] 
    return xsamples
```

```python
x = arange(-.2,.201,.001) 
a = 10e-6
d = 15e-6 
l = 1 
lambd = 632.8e-9 
intensity = lambda x: (sinc(a*x/(lambd*l))**2)*(cos(2*pi*d*x/(lambd*l))**2)
Fxin = distribute1D(x,intensity(x),10000)
```

```python
n, bins, patches = hist(Fxin,50,normed=True,label="Sampled data")
plot(x,intensity(x)*max(n),'r-',label='True distribution')
xlabel("x (m)")
ylabel("double-slit intensity")
xlim(-0.2,0.2)
show()
```
