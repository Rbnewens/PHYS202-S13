```python
#some sample data for plots
y = arange(-5., 5.+.01, .01)
E = cosh(y)
mp = sinh(y)

#object oriented setup
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

#main plot parameters go here
plt.plot(y, E, linestyle = '-', color = 'red', linewidth = 1., label = "$foo$")
plt.plot(y, mp, label = "$bar$")

#title
ax.set_title('stub title')

#spines, aka axis
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_color('none')

#tick marks
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#axis labels
plt.xlabel("$stub$")
plt.ylabel("$stub()$")

#legend location
plt.legend(loc='best', fontsize = 9)

#show the plot. ties focus outside of ipython.
plt.show()
```

Subplots

```python
x = arange(0, 7, 0.01)

subplot(2, 1, 1)
plot(x, sin(x))

subplot(2, 2, 3)
plot(x, cos(x))

subplot(2, 2, 4)
plot(x, sin(x)*cos(x))
```
