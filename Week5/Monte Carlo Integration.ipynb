```python
f = lambda x: x**(0.5) + sin(x)
```

```python
x = arange(0., 15., 0.001)
```

```python
myplot = plot(x,f(x),lw=3)
ylabel(r'$x^{1/2} + \sin{x}$',fontsize=20)
xlabel(r'$x$',fontsize=20)
xlim(0.,15.)
show()
```

```python
xmin = 1.7
xmax = 12.4

#find the maximum value of f within the boundary to set ymax
subx = x[logical_and((x > xmin),(x < xmax))]
ymax = max(f(subx))*1.05
ymin = 0

#replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

#plot the box
plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')#veritcal left
plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')#veritcal right
plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')#horizontal top

show()
```

```python
#random_sample?

N = 10000
samples_x = xmin + (xmax-xmin)*random_sample(N);
samples_y = ymin + (ymax-ymin)*random_sample(N);

#replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:200],samples_y[1:200],'.')

#plot the box
plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')#veritcal left
plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')#veritcal right
plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')#horizontal top

show()
```

try a mask on samples_y

```python
newmask = (samples_y < f(samples_x))
```

```python
sum(newmask)
```

replot the points that lie below the function for proof:

```python
#replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:200],samples_y[1:200],'.')

#show the ones below the function in red
plot(samples_x[newmask[:200]],samples_y[newmask[:200]], 'r.')

#plot the box
plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')#veritcal left
plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')#veritcal right
plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')#horizontal top

show()
```

```python
#compute integral stepV: area of rectagnle times ratio of # of red and green dots

I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I
```

```python
#exact (indefinite) analytical solution to our integral:

Iana = lambda x: (2./3.)*x**(3/2.) - cos(x)
Iexact = Iana(xmax)-Iana(xmin)
Ierr = 100*(Iexact-I)/Iexact
print "I_exact =",Iexact
print "I_hitormiss =",I
print "Error on I =",Ierr,"%"
```

```python
#sample mean method
f = lambda x: x**(.5) + sin(x)
x = arange(0.,15., .001)

#integration limits
xmin = 1.7
xmax = 12.4

#bounding box y limits
subx = x[logical_and((x > xmin), (x < xmax))]
ymax = max(f(subx))*1.05
ymin = 0.

N = float(1E4)
samples_x = xmin + (xmax-xmin)*random_sample(N)
guess_y = f(samples_x)
avg = np.average(guess_y)
print "average value is", avg
I = avg*(xmax-xmin)
print "I_MC =", I


Iana  = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Iana(xmax) - Iana(xmin)
Ierr = 100*(Iexact - I) / Iexact
print "I_exact =", Iexact
print "I_MC =", I
print "Error of MC =", Ierr, "%"
```

```python
avgraph = np.ones(x.size)
avgraph *= avg

myplot = plot(x, f(x), lw=3)
plot(x,avgraph)
ylabel(r'$x^{1/2} + \sin(x)$', fontsize=20)
xlabel(r"$x$", fontsize = 20)
xlim(0., 15.)
show()
```
